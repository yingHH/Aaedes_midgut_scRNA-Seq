# plot umap

```python
umap = pd.read_csv('../replot/umap.csv', index_col=0)
umap = umap.loc[info.index]
umap
```

```python
cell_id_dct = {
'EC-3': 0,
'ISC/EB': 1,
'EC-2': 2,
'ISC/EB-d1': 3,
'ISC/EB-d2': 4,
'EC-AA': 5,
'EC-1': 6,
'EC-im': 7,
'ISC/EB-q': 8,
'EC-trypsin 5G1': 9
}
```

```python
id_cell_dct = {v:k  for k, v in cell_id_dct.items()}
id_cell_dct
```

```python
lin = [
[1, 4, 2, 9, 6, 7],
[1, 4, 2, 9, 6, 5],
[1, 4, 2, 0],
[1, 4, 3, 8],
]
```

```python
lins = [[id_cell_dct[i] for i in l] for l in lin]
lins
```

```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
import seaborn as sns

fig, ax = plt.subplots(figsize=(6, 6))

c = sns.color_palette('deep')
set1 = sns.color_palette('Set1')

grp_color = {
    'ISC/EB': c[4],
    'ISC/EB-q': c[5],
    'ISC/EB-d1': c[6],
    'ISC/EB-d2': set1[1],
    'EC-AA': c[-2],
    'EC-im': c[1],
    'EC-1': c[2],
    'EC-trypsin 5G1': c[0],
    'EC-2': c[3],
    'EC-3': set1[3],
}

lb_loc = {}
for lb in grp_color.keys():
    pdf = umap[info['celltype'] == lb]
    l1, = ax.plot(
        pdf['umap1'], pdf['umap2'], '.', color=grp_color[lb],
        alpha=0.6, ms=2.5) #e7e7e7

    x, y = np.mean(pdf[['umap1', 'umap2']].values, axis=0)
    lb_loc[lb] = (x, y)

print('plot across')
for lin in lins:
    last_p = ''
    for i, lb in enumerate(lin):
        p = lb_loc[lb]
        if i > 0:
            x, y = p
            xl, yl = last_p
            color = 'black'

            if lb == 'EC-S':
                x = x + 0.5
                y = y -2
                _ = ax.plot((xl, x), (yl, y), '--', color=color)
                ax.arrow(
                    xl + (x - xl) * 0.35, yl + (y - yl) * 0.35, (x - xl) * 0.01, (y - yl) * 0.01, 
                    width=0.01, head_width=0.4, overhang=0.3,
                    color='k', zorder=10)
                last_p = (x, y)
            elif lb == 'EC-Trypsin 5G1':
                x = x + 1
                y = y -0.5
                _ = ax.plot((xl, x), (yl, y), '--', color=color)
                ax.arrow(
                    xl + (x - xl) * 0.35, yl + (y - yl) * 0.35, (x - xl) * 0.01, (y - yl) * 0.01, 
                    width=0.01, head_width=0.4, overhang=0.3,
                    color='k', zorder=10)
                last_p = (x, y)
            else:
                _ = ax.plot((xl, x), (yl, y), '--', color=color)
                ax.arrow(
                    xl + (x - xl) * 0.35, yl + (y - yl) * 0.35, (x - xl) * 0.01, (y - yl) * 0.01, 
                    width=0.01, head_width=0.4, overhang=0.3,
                    color='k', zorder=10)
                last_p = (p)
        else:
            last_p = p

print('plot text')
for lb in grp_color.keys():
    x, y = lb_loc[lb]  
    if lb == 'EC-AA':
        txt = ax.text(x - 1.5, y, lb, ha='center', va='center')
        txt.set_path_effects([
            PathEffects.withStroke(linewidth=2, foreground='w')])
    else:
        txt = ax.text(x, y, lb, ha='center', va='center')
        txt.set_path_effects([
            PathEffects.withStroke(linewidth=2, foreground='w')])
    

# pdf = umap[
#     umap.index.isin(adata_sgl.obs.index)
# ][adata_sgl.obs['sp'] == 'midgut 10d']
# l2, = ax.plot(pdf['umap1'], pdf['umap2'], '.', color=c[3], alpha=0.6, ms=2.5)

ax.arrow(
    -10, -10, 0, 3.5, 
    width=0.01, head_width=0.2, overhang=0.3, 
    color='k', zorder=10)
ax.arrow(
    -10, -10, 3, 0, 
    width=0.01, head_width=0.2, overhang=0.3,
    color='k', zorder=10)

ax.set_axis_off()

ax.text(-8.4, -10.1, 'UMAP-1', ha='center', va='top')
ax.text(-10.1, -8.1, 'UMAP-2', ha='right', va='center', rotation=90)

ax.set_aspect('equal')
ax.set_xlim(-11.52392974688526, 6.038771096925777)
ax.set_ylim(-11.204441909223405, 6)
# ax.plot(-12, 9, '.', color=c[0], mec=c[0], ms=11, alpha=0.9)
# ax.text(-11.5, 9, 'midgut 10d', ha='left', va='center')

plt.savefig('./plots/f3.trajectory.png', dpi=600, bbox_inches='tight')
```

```python
diff_early = pd.read_csv('./diff.early.csv', index_col=0)
diff_early = diff_early.merge(gene_des, left_on='feature_id', right_on='gene')
diff_early = diff_early[diff_early['description'] != 'unspecified product']
diff_early = diff_early[diff_early['pvalue'] <= 0.05]
diff_early = diff_early.sort_values(['waldStat'], ascending=True)

diff_early
```
