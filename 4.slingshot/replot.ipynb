# plot umap

```python
umap = pd.read_csv('../replot/umap.csv', index_col=0)
umap = umap.loc[info.index]
umap
```

```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects
import seaborn as sns

fig, ax = plt.subplots(figsize=(6, 6))

c = sns.color_palette('deep')
set1 = sns.color_palette('Set1')

grp_color = {
    'ISC/EB-p': c[4],
    'ISC/EB-m': c[5],
    'ISC/EB-d': c[6],
    'EC-AA': c[-2],
    'EC-Na': c[1],
    'EC': c[2],
    'EC-Trypsin 5G1': c[0],
    'EC-S': c[3],
}

lb_loc = {}
for lb in grp_color.keys():
    pdf = umap[info['celltype'] == lb]
    l1, = ax.plot(
        pdf['umap1'], pdf['umap2'], '.', color=grp_color[lb],
        alpha=0.6, ms=2.5) #e7e7e7

    x, y = np.mean(pdf[['umap1', 'umap2']].values, axis=0)
    lb_loc[lb] = (x, y)

lin1 = ['ISC/EB-p', 'ISC/EB-m', 'ISC/EB-d', 'EC', 'EC-Trypsin 5G1', 'EC-Na', 'EC-AA']
lin2 = ['ISC/EB-p', 'ISC/EB-m', 'ISC/EB-d', 'EC', 'EC-S']

print('plot across')
for lin in [lin1, lin2]:
    last_p = ''
    for i, lb in enumerate(lin):
        p = lb_loc[lb]
        if i > 0:
            x, y = p
            xl, yl = last_p
            color = 'black'

            if lb == 'EC-S':
                x = x + 0.5
                y = y -2
                _ = ax.plot((xl, x), (yl, y), '--', color=color)
                ax.arrow(
                    xl + (x - xl) * 0.35, yl + (y - yl) * 0.35, (x - xl) * 0.01, (y - yl) * 0.01, 
                    width=0.01, head_width=0.4, overhang=0.3,
                    color='k', zorder=10)
                last_p = (x, y)
            elif lb == 'EC-Trypsin 5G1':
                x = x + 1
                y = y -0.5
                _ = ax.plot((xl, x), (yl, y), '--', color=color)
                ax.arrow(
                    xl + (x - xl) * 0.35, yl + (y - yl) * 0.35, (x - xl) * 0.01, (y - yl) * 0.01, 
                    width=0.01, head_width=0.4, overhang=0.3,
                    color='k', zorder=10)
                last_p = (x, y)
            else:
                _ = ax.plot((xl, x), (yl, y), '--', color=color)
                ax.arrow(
                    xl + (x - xl) * 0.35, yl + (y - yl) * 0.35, (x - xl) * 0.01, (y - yl) * 0.01, 
                    width=0.01, head_width=0.4, overhang=0.3,
                    color='k', zorder=10)
                last_p = (p)
        else:
            last_p = p

print('plot text')
for lb in grp_color.keys():
    x, y = lb_loc[lb]  
    if lb == 'EC-S':
        txt = ax.text(x + 1, y - 2, lb, ha='center', va='center', zorder=11)
        txt.set_path_effects([
            PathEffects.withStroke(linewidth=2, foreground='w')])
    elif lb == 'Unkn':
        txt = ax.text(x + 1.25, y, lb, ha='center', va='center', zorder=11)
        txt.set_path_effects([
            PathEffects.withStroke(linewidth=2, foreground='w')])
    elif lb == 'EC-AA':
        txt = ax.text(x + 1, y - 1, lb, ha='center', va='center', zorder=11)
        txt.set_path_effects([
            PathEffects.withStroke(linewidth=2, foreground='w')])
    elif lb == 'EC-Trypsin 5G1':
        txt = ax.text(x + 1.5, y - 1, lb, ha='center', va='center', zorder=11)
        txt.set_path_effects([
            PathEffects.withStroke(linewidth=2, foreground='w')])
    elif lb == 'ISC/EB-p':
        txt = ax.text(x + 1, y, lb, ha='center', va='center', zorder=11)
        txt.set_path_effects([
            PathEffects.withStroke(linewidth=2, foreground='w')])
    else:
        txt = ax.text(x, y, lb, ha='center', va='center', zorder=11)
        txt.set_path_effects([
            PathEffects.withStroke(linewidth=2, foreground='w')])
    

# pdf = umap[
#     umap.index.isin(adata_sgl.obs.index)
# ][adata_sgl.obs['sp'] == 'midgut 10d']
# l2, = ax.plot(pdf['umap1'], pdf['umap2'], '.', color=c[3], alpha=0.6, ms=2.5)

ax.arrow(
    -11, -8, 0, 3, 
    width=0.01, head_width=0.2, overhang=0.3, 
    color='k', zorder=10)
ax.arrow(
    -11, -8, 3, 0, 
    width=0.01, head_width=0.2, overhang=0.3,
    color='k', zorder=10)

ax.set_axis_off()

ax.text(-9.4, -8.1, 'UMAP-1', ha='center', va='top')
ax.text(-11.1, -6.3, 'UMAP-2', ha='right', va='center', rotation=90)

ax.set_aspect('equal')

# ax.plot(-12, 9, '.', color=c[0], mec=c[0], ms=11, alpha=0.9)
# ax.text(-11.5, 9, 'midgut 10d', ha='left', va='center')

plt.savefig('./plots/f3.trajectory.png', dpi=600, bbox_inches='tight')
```
