```python
import pandas as pd
```

```python
EC = {8: 'ISC/EB',
 9: 'ISC/EB-q',
 4: 'ISC/EB-d1',
 5: 'ISC/EB-d2',
 7: 'EC-im',
 3: 'EC-1',
 2: 'EC-2',
 0: 'EC-3',
 1: 'EC-trypsin 5G1',
 11: 'EC-AA'}
```

```python
re_EC = {k: i for i, k in enumerate(set(EC.values()))}
re_EC
```

```python
df = pd.read_csv('./seurat.meta_data.csv', index_col=0)
df['celltype'] = df['seurat_clusters'].map(lambda x: EC[x])
df['celltype_id'] = df['celltype'].map(lambda x: re_EC[x])
df
```
